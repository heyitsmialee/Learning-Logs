## 1. 원 핫 인코딩(One-Hot Encoding)
범주형, 명목형 변수의 각 범주를 독립적인 이진 특성으로 변환하는 기법

```python
import pandas as pd

# 타이타닉 데이터셋 로드
titanic = pd.read_csv('./titanic.csv')
```

```python
titanic.info()
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

train_ex1 = titanic.copy()

encoder = OneHotEncoder(sparse_output = False)
train_encoded = encoder.fit_transform(train_ex1[['Sex']])

encoded_columns = encoder.get_feature_names_out(['Sex'])
train_encoded_df = pd.DataFrame(train_encoded, columns=encoded_columns)

train_ex1 = pd.concat([train_ex1, train_encoded_df], axis=1).drop(['Sex'], axis=1)

display(encoded_columns)
display(train_ex1.head())
```

## 2. 다수의 고유 범주를 가지는 명목형 변수의 인코딩

### 2-1. 바이너리 인코딩(Binary Encoding)

고유 범주의 갯수가 많은 명목형 변수를 인코딩할 수 있는 대표적인 기법이 바이너리 엔코딩(binary encoding)입니다.
이 기법은 데이터의 차원을 효과적으로 줄이면서 각 범주를 명확하게 표현할 수 있는 장점을 가지고 있습니다.

바이너리 인코딩은 각 범주를 레이블 인코딩으로 변환한 후, 이진수로 표현하여 범주의 수에 비례하지 않는 고정된 길이의 열을 생성합니다. 이는 차원의 증가를 상당히 제한하며, 동시에 각 범주를 효율적으로 표현할 수 있게 합니다.

```python
import category_encoders as ce

train_ex5 = titanic.copy()

train_ex5['Embarked'].fillna('Unknown', inplace=True)

# Creating a BinaryEncoder instance
encoder_ex5 = ce.BinaryEncoder(cols=['Embarked'])

# Fitting and transforming the 'Embarked' column
train_encoded_ex5 = encoder_ex5.fit_transform(train_ex5['Embarked'])

encoded_columns_ex5 = train_encoded_ex5.columns

# Creating a DataFrame from the binary encoded data
train_encoded_df5 = pd.DataFrame(train_encoded_ex5, columns=encoded_columns_ex5)
train_ex5 = pd.concat([train_ex5, train_encoded_df5], axis=1).drop(['Embarked'], axis=1)

display(encoded_columns_ex5)
display(train_ex5.head())
```
