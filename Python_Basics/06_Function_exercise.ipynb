### [함수를 사용하는 이유]

- 연관된 작업을 묶음으로 관리하고 언제든 재활용할 수 있다
- 연관된 작업들을 모아서 더 복잡한 작업들을 구성한다.

  예)평균계산기=합계산+원소개수+나눗셈

### [함수] : 할 일(=작업)들의 묶음

1. def : define. 함수를 정의한다는 키워드
2. 이름: 함수를 사용(호출)하려면 부를 이름이 필요하다.
3. 입력 값: 인자(argument)라고 부른다.
4. 몸통(body): 함수의 동작 내용. = 함수의 코드 블록.
5. 반환 값: 함수의 결과 값. return 키워드로 결과를 내어준다.

인자와 반환값은 생략이 가능하다.
- now같은 함수는 입력값이 없음.

```python
def calcSum(inputlist):
  sum = 0
  for input in inputlist:
    sum = sum + input
  return sum

sample = [10, 20, 30, 40, 50]

sum = calcSum(sample)
print(sum)
```

함수 이름을 호출하면 함수 내용이 실행된다

• 함수 호출: 예) calcSum(numOfCustomers)

• 인자 값 전달: 예) calcSum(numOfCustomers)

• 반환 값 저장: 예) sumOfCustomers = calcSum(numOfCustomers)
 => 반환값:함수의실행결과

```python
# 나눗셈을 계산하는 함수를 만드는 방법

def calcDivision(value, count):
  if count == 0:
    div = None
  else:
    div = value/count

  return div

# 함수를 호출하기

numOfCustomers = [1, 2, 3, 4, 5]

sumOfCustomers = calcSum(numOfCustomers)
print(sumOfCustomers)
```

```python
# 평균을 계산하는 함수를 만드는 방법

def calcLen(samples):
  length = len(samples)
  return length

def calcAverage(samples):
  # 합을 계산한다.
  sum = calcSum(samples)
  # 원소의 개수를 구한다.
  length = calcLen(samples)
  # 결과를 알려준다.
  avg = sum / length

  return avg

sample = [1,2,3,4,5]
calcAverage(sample)
```

## 프로그램과 함수의 관계

함수는 재활용이 가능한데, 입력값과 출력값만 알면 되기 때문에 함수 내에 있는 변수가 프로그램에 있는 변수가 서로 독립적으로 작용한다.

함수 안에 프로그래밍 변수와 같은 변수가 있어도, 함수 안에 있는 값을 프로그램이 변수에 넣어주지는 않는다.

*   전역변수(global) : 프로그램은 함수 안의 변수를 보지 않는다.
*   자역변수(local) : 함수는 자신의 코드블록 밖의 변수는 보지 못한다.

=> 지역변수와 전역변수의 이름은 같게 설정해도 무관하다.


image.png

```python
# 지역변수와 전역변수의 이름은 같게 설정해도 무관하다. (서로의 존재를 알지 못함)

sum = -1 #전역변수(Global variable)

def calcSum(inputlist):
  global sum
  sum = 0   #지역변수(local variable)

  for item in inputlist:
    sum = sum + item

  return sum

samples = [1, 3, 5, 7, 9, 11]

sum = calcSum(samples) #전역변수가 업데이트 됨

print(f"sum = {sum}")
```

- 함수 만으로는 동작이 일어나지 않고, 함수를 호출해서 사용해야 동작을 한다.

2. 함수의 구성 요소
- 이름, 입력인자, 반환 값 + def, 함수 몸통(Body)

3. 전역변수 VS 지역변수
- 전역변수 : 프로그램 전체에 걸쳐 사용되고 관리되는 값
- 지역변수 : 함수 안에서만 사용되는 값
- 서로 모르기 때문에 이름을 같이 써도 된다.
- 자동으로 업데이트 되지 않기 때문에 내부에서 지역변수를 부를 때 불러서 써야함 (시험에서 유의해야할 듯)

```python
#######################################################################################
#
#   1. 두 수를 입력으로 받고 이 중 큰 수를 반환하는 기능을 제공하는 함수를 작성하시오.
#
#######################################################################################

def getLargerOne(a, b):
    output = None #지역변수의 계산 결과를 저장할 곳. 아직 값이 없음 (None 값임)

    if a >= b:
      output = a

    else:
      output = b

    return output

num1 = 5
num2 = 10

print(getLargerOne(num1, num2)) #함수의 호출(실행) 결과 확인
```

```python
#######################################################################################
#
#   2. 아래 calcSum() 함수는 주어진 리스트의 합(sum)을 구하는 함수입니다.
#   이 함수를 이용해서 sample 리스트의 평균을 구하는 함수 calcAverage()를 완성하시오.
#
#######################################################################################

def calcSum(inputList):
    sum = 0

    for item in inputList:
        sum = sum + item #기존 sum 변수의 값에 리스트에서 가져온 원소인 item 값을 더함 -> 다시 sum 변수에 저장

    return sum

sample = [-8, 2, 12, 17, 21, 25, 28, 23, 19, 12, 8, -5]

def calcAverage(inputList):
    avg = None #계산 결과가 아직 없기 때문에 계산 결과의 초기 값 = None

    #   함수를 완성 하시오.

    # 합을 구한다
    sum = calcSum(inputList)

    # 개수를 구한다
    length = len(inputList)

    # 평균을 구한다
    if length > 0:    # 리스트의 원소가 1개 이상인 경우에만 평균이 존재함
      avg = sum / length

    else:
      avg = None

    return avg

result = calcAverage(sample)

print(round(result,2))
```

```python
#######################################################################################
#
#   3. 아래는 5개 도시의 월 평균 기온입니다.
#   도시 각각의 연평균 기온을 구한 후 cityTempAvg 리스트에 저장하시오.
#
#######################################################################################

cityTempDict = {}   #   5개 도시의 월 평균 기온들을 보관하는 딕셔너리

cityTempDict['서울'] = [-8, -4, 7, 13, 21, 29, 32, 33, 27, 19, 8, -5]
cityTempDict['춘천'] = [-14, -9, 3, 9, 19, 26, 31, 32, 24, 13, 2, -9]
cityTempDict['청주'] = [-5, -3, 9, 15, 23, 28, 30, 32, 29, 20, 12, -4]
cityTempDict['광주'] = [-5, -1, 8, 17, 23, 29, 32, 33, 25, 15, 6, -3]
cityTempDict['포항'] = [-4, 3, 12, 16, 24, 30, 33, 34, 28, 22, 11, -3]

cityTempAvg = []    #   5개 도시 각각의 연 평균 기온을 저장할 리스트

#   각 도시의 연 평균 기온을 구한 후 cityTempAvg 리스트에 저장하시오.
for key in cityTempDict.keys():

    # 각 도시의 월 평균 기온 리스트를 불러온다.
    templist = cityTempDict[key]

    # 월 평균 기온 리스트의 평균을 구한다.
    avgtemp = calcAverage(templist) # 도시의 연 평균 기온(=월 평균 기온의 평균) 계산

    # 평균값을 cityTempAvg 리스트에 저장한다.
    cityTempAvg.append(round(avgtemp,2))  # 계산된 값을 반올림하여 소수 2째자리까지 남긴다

print(cityTempAvg)  # 결과 확인
```

```python
#######################################################################################
#
#   4. 문제 3에서 구한 연 평균 기온(cityTempAvg) 중 최대값을 찾는 프로그램을
#   max() 함수를 사용하지 말고 작성하시오.
#
#######################################################################################

maxVal = cityTempAvg[0] # 수(원소)를 하나 선택한 후 보관한다.

# 리스트에서 값 하나를 뽑는다.
for item in cityTempAvg:
    # 뽑은 값을 다른 값들과 비교한다
    if item > maxVal:
      # 더 큰 값이 나오면 그 값을 저장한다.
      maxVal = item

    # 다른 값과 비교해서 저장 값이 더 크면 다음 값과 계속 비교한다.
    else:
      continue
      # pass : 해야할 내용이 없으므로 pass라고 해도 무관함.

print(round(maxVal,2))   # 결과 확인

# 리스트의 마지막 원소까지 2~4 과정을 반복한 후, 보관하고 있는 원소가 리스트에서의 최대값이 된다.
```

```python
#######################################################################################
#
#   5. 입력으로 주어진 리스트에서 최대값을 찾는 함수를 max() 함수를 사용하지 말고 작성하시오.
#
#######################################################################################

def getMaxValue(inputList):

    maxVal = inputList[0] # 수(원소)를 하나 선택한 후 보관한다.

    #   입력으로 주어진 리스트에서 최대값을 찾는 코드를 작성하시오.
    for input in inputList:
      if input > maxVal :
        maxVal = input
      else:
        continue

    return maxVal

print(getMaxValue(cityTempAvg))
```
